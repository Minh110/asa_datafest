```python
play_start = int(timelogs.iloc[0])
play_end = 0
total = 0
print("Real time: " + str(int(timelogs.iloc[len(timelogs) - 1]) / 3600))

for i in range(1, len(timelogs)):
  row = int(timelogs.iloc[i])
  play_end = row

  if i == (len(timelogs) - 1): 
    total = total + (play_end - play_start)
    #print("Ending calculate: " + str(play_end - play_start))
    break 

  row1 = int(timelogs.iloc[i + 1])

  if ((row1 - play_end) / 3600) > 12:
    total = total + (play_end - play_start)
    #print("Calculate: " + str(play_end - play_start))
    play_start = row1

print("Actual playing time: " + str(total/3600) + " hours")
```
