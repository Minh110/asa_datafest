```python
def timeCal(df):
  play_start = int(df[0])
  play_end = 0
  total = 0
  #print("Real time: " + str(int(timelogs.iloc[len(timelogs) - 1]) / 3600))

  for i in range(1, len(df)):
    row = int(df[i])
    play_end = row

    if i == (len(df) - 1): 
      total = total + (play_end - play_start)
      #print("Ending calculate: " + str(play_end - play_start))
      break 

    row1 = int(df[i + 1])

    if ((row1 - play_end) / 3600) > 12:
      total = total + (play_end - play_start)
      #print("Calculate: " + str(play_end - play_start))
      play_start = row1
  
  return total

  #print("Actual playing time: " + str(total/3600) + " hours")
```

```python
big_timelogs["actual"] = 0
#big_timelogs
```

```python
for i in range(len(big_timelogs)): 
  actualTime = timeCal(big_timelogs.iloc[i][1])
  #print(actualTime/3600)
  big_timelogs.at[i, "actual"] = actualTime/3600 

big_timelogs
```
